```python
# Making the imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-100, 100.0, 0.1)
y = np.arange(-100, 100.0, 0.1)
```

```python
def ellipse(x, y, a, b):
    """Function that plots an ellipse
    
    Input: x(x-coordinate), y(y-coordinate), a(width), b(height)
    Output: plot   """
    
    x,y = np.meshgrid(x,y)
    
    z = (( x**2 / a**2 ) + ( y**2 / b**2 ) - 1 ) 
    
    fig, ax = plt.subplots()
    
    ax.contour(x,y,z,[0])
   # ax.set_aspect(1)
    plt.xlim(-15,15)
    plt.ylim(-25,25)
    
    return plt.show()
```

```python
a = 12
b = 2
print(ellipse(x, y, a, b))
```

```python
a = 4
b = 20
print(ellipse(x, y, a, b))
```
