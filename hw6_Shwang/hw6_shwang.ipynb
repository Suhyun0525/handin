```python
def quad( a ,b ,c ):
    from math import sqrt
    """ This is the function to calculate the roots of a quadratic equation.
    a, b, and c must be coefficients of each variables in a quadratic equation. """
    
    f1 = b**2 - 4*a*c
    f2 = sqrt(f1)
    positive = ( -b + f2 ) / ( 2 * a )
    negative = ( -b - f2 ) / ( 2 * a )
    return positive, negative
```

```python
"""Since our equation is 𝑥**2-x-12 = 0, a = 1, b = -1, c = -12"""
roots = quad( 1 ,-1 ,-12 )
print( roots )
```
